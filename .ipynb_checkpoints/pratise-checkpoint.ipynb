```python
a=np.array([["num", 2,3,2],[2,3,4,2]],dtype=object) ##help to write in python datatypes   
print(a)
d1=a.flatten() ##converts a multi-dimensional NumPy array into a 1-D array.
d=np.array([x for x in d1 if isinstance(x, (int, float))])  ##Checks whether a value belongs to a specified data type.what type a value is.
print(d)
```

```python
r=np.array([1,2,3,4])
t=np.array([[1,2,4,3],[2,6,3,2]])
print(t ,"\n",r) # for the new line excute the output
```

```python
r=np.array([1,2,3,4])
t=np.array([[1,2,4,3],[2,6,3,2]])
r1=np.array([[[2,3,2,5],[5,3,2,2],[2,1,4,2]]])
print(t ,"\n",r ,"\n",r1)
p=np.combined(r,t,r1)  # combined the element
print(p)
```
